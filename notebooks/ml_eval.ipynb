### Importing Libraries

```python
import joblib 
from datasets import load_dataset
from sklearn.metrics import accuracy_score
```

```python
from src.ml_utils import clean_text
```

### Load ML model

```python
def load_ml_model(model_name):
    model_path = f"../models/ml/{model_name}.joblib"
    model = joblib.load(model_path)
    return model
```

### Test set

```python
dataset = load_dataset('imdb', split='test')
texts = dataset["text"]
labels = dataset["label"]

# clean text
X_test = [clean_text(x) for x in texts]
y_test = list(labels)
```

### Evaluation

```python
models = [
    "logistic_regression",
    "linear_svc",
    "multinomial_nb"
]

for name in models:
    model = load_ml_model(name)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    print(f"{name}: {acc:.4f}")
```
